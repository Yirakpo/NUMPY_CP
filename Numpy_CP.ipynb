```python


import numpy as np

# Create the grades array
grades = np.array([85, 90, 88, 92, 95, 80, 75, 98, 89, 83])
print(f"Grades: {grades}")

# Calculate the mean of the grades
mean_grades = np.mean(grades)
print(f"Grades: {grades}")

# Calculate the median of the grades
median_grades = np.median(grades)
print(f"Mean of the grades: {mean_grades}")

# Calculate the standard deviation of the grades
std_deviation = np.std(grades)
print(f"Standard Deviation of the grades: {std_deviation}")

# Find the maximum grade
max_grade = np.max(grades)
print(f"Maximum grade: {max_grade}")

# Find the minimum grade
min_grade = np.min(grades)
print(f"Minimum grade: {min_grade}")

# Sort the grades in ascending order
sorted_grades = np.sort(grades)
print(f"Sorted grades: {sorted_grades}")

# Find the index of the highest grade
index_max_grade = np.argmax(grades)
print(f"Index of the highest grade: {index_max_grade}")

# Count the number of students who scored above 90
count_above_90 = np.sum(grades > 90)
print(f"Number of students who scored above 90: {count_above_90}")

# Calculate the percentage of students who scored above 90
percentage_above_90 = np.mean(grades > 90) * 100
print(f"Percentage of students who scored above 90: {percentage_above_90:.2f}%")

# Calculate the percentage of students who scored below 75
percentage_below_75 = np.mean(grades < 75) * 100
print(f"Percentage of students who scored below 75: {percentage_below_75:.2f}%")

# Extract all the grades above 90 and put them in a new array called "high_performers"
high_performers = grades[grades > 90]
print(f"Grades of high performers (above 90): {high_performers}")

# Create a new array called "passing_grades" that contains all the grades above 75
passing_grades = grades[grades > 75]
print(f"Passing grades (above 75): {passing_grades}")







```
